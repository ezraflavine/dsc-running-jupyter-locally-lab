```python
#when looking for data from scratch we use the random.
import random
import datetime
```

```python
random.seed(3)
data = [['Name', 'DOB', 'Grade', 'Attitude']] # list to store the column names
#loop to generate data
for i in range(150):
    name = "student" + str(i+1)
    dob = datetime.data(random.randint(1980, 2000), random.randint(1, 12),
                        random.randint(1, 28)).strftime("%d/%m/%y")
    grade = random.randint(0, 100)
    attitude = random.choice(['positive', 'Negative'])
    data.append([name, dob, grade, attitude])

print(data)
```
